## AutoEncoder

데이터에 대한 효율적인 압축을 신경망을 통해 자동으로 학습하는 모델. 기본적인 형태는 인코더 디코더로 이루어져있고, 입력을 x로 주어졌을때, 인코터를 거치면서 z(letent vector)가 되고 이를 디코더를 거치면서 x'(x가 복원된 형태)가 된다. 오토인코더는 비지도학습으로 분류되는데, x가 정답 레이블로써 활용되기 때문이다. loss는 x와 x'의 차이로 계산된다.(보통은 픽셀단위)


```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import torchvision.datasets as dset
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
```

```python
batch_size = 256
learning_rate = 0.001
num_epoch = 10
```

```python
mnist_train = dset.MNIST("./", train=True, transform=transforms.ToTensor(), target_transform=None, download=True)
mnist_test = dset.MNIST("./", train=False, transform=transforms.ToTensor(), target_transform=None, download=True)
```

```python
train_loader= DataLoader(mnist_train, batch_size=batch_size, shuffle=True, num_workers=2, drop_last=True)
test_loader=DataLoader(mnist_test, batch_size=batch_size, shuffle=False, num_workers=2, drop_last=True)
```

```python
for index, [image, label] in enumerate(train_loader):
    x = image
    y = label
    print(x.shape)
    break
```

```python
#기본적인 AutoEncoder
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder,self).__init__()
        self.encoder = nn.Linear(28*28,20)
        self.decoder = nn.Linear(20, 28*28)
    def forward(self,x):
        x = x.view(batch_size, -1)
        encoded = self.encoder(x)
        out = self.decoder(encoded).view(batch_size,1,28,28)
        return out
         
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = Autoencoder().to(device)
loss_func = nn.MSELoss().to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
loss_arr = []
for i in range(num_epoch):
    for j, [image, label] in enumerate(train_loader):
        image = image.to(device)
        optimizer.zero_grad()
        output = model(image)
        loss = loss_func(output,image) #입력이 레이블으로 활용된다.
        loss.backward()
        optimizer.step()
    
    
    print(f'epoch{i+1}:{loss}')
    loss_arr.append(loss.cpu().data.numpy())
```

```python
with torch.no_grad():
    for j, [image, label] in enumerate(test_loader):
        image = image.to(device)
        output = model(image)
        loss = loss_func(output,image) #입력이 레이블으로 활용된다.

print(f'test_loss:{loss}') 
```

```python
#Convolution AutoEncoder-Encoder
class Encoder(nn.Module):
    def __init__(self):
        super(Encoder,self).__init__()
        self.layer1 = nn.Sequential(
                        nn.Conv2d(1,16,3,padding=1),                            # batch x 16 x 28 x 28
                        nn.ReLU(),
                        nn.BatchNorm2d(16),
                        nn.Conv2d(16,32,3,padding=1),                           # batch x 32 x 28 x 28
                        nn.ReLU(),
                        nn.BatchNorm2d(32),
                        nn.Conv2d(32,64,3,padding=1),                           # batch x 32 x 28 x 28
                        nn.ReLU(),
                        nn.BatchNorm2d(64),
                        nn.MaxPool2d(2,2)                                       # batch x 64 x 14 x 14
        )
        self.layer2 = nn.Sequential(
                        nn.Conv2d(64,128,3,padding=1),                          # batch x 64 x 14 x 14
                        nn.ReLU(),
                        nn.BatchNorm2d(128),
                        nn.MaxPool2d(2,2),
                        nn.Conv2d(128,256,3,padding=1),                         # batch x 64 x 7 x 7
                        nn.ReLU()
        )
        
                
    def forward(self,x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = out.view(batch_size, -1)
        return out
```

```python
class Decoder(nn.Module):
    def __init__(self):
        super(Decoder,self).__init__()
        self.layer1 = nn.Sequential(
                        nn.ConvTranspose2d(256,128,3,2,1,1),                    # batch x 128 x 14 x 14
                        nn.ReLU(),
                        nn.BatchNorm2d(128),
                        nn.ConvTranspose2d(128,64,3,1,1),                       # batch x 64 x 14 x 14
                        nn.ReLU(),
                        nn.BatchNorm2d(64)
        )
        self.layer2 = nn.Sequential(
                        nn.ConvTranspose2d(64,16,3,1,1),                        # batch x 16 x 14 x 14
                        nn.ReLU(),
                        nn.BatchNorm2d(16),
                        nn.ConvTranspose2d(16,1,3,2,1,1),                       # batch x 1 x 28 x 28
                        nn.ReLU()
        )
        
    def forward(self,x):
        out = x.view(batch_size,256,7,7)
        out = self.layer1(out)
        out = self.layer2(out)
        return out

```

```python
class Conv_AutoEncoder(nn.Module):
    def __init__(self):
        super(Conv_AutoEncoder,self).__init__()
        self.encoder = Encoder()
        self.decoder = Decoder()
    
    def forward(self, x):
        out = self.encoder(x)
        out = self.decoder(out)
        return out
    
```

```python
#훈련방식1
model = Conv_AutoEncoder().to(device)
loss_func = nn.MSELoss().to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
#훈련방식2
encoder = Encoder().to(device)
decoder = Decoder().to(device)

# 인코더 디코더의 파라미터를 동시에 학습시키기 위해 이를 묶는 방법입니다.
parameters = list(encoder.parameters())+ list(decoder.parameters())

loss_func = nn.MSELoss()
optimizer = torch.optim.Adam(parameters, lr=learning_rate)
```

```python
#훈련방식2
loss_arr = []
for i in range(num_epoch):
    for j,[image,label] in enumerate(train_loader):
        optimizer.zero_grad()
        image = image.to(device)
        output = encoder(image)
        output = decoder(output)
        loss = loss_func(output,image)
        loss.backward()
        optimizer.step()
    
    
    print(f'epoch{i+1}:{loss}')
    loss_arr.append(loss.cpu().data.numpy())
```

## Semantic Segmentation

-> UNet을 대표적으로 사용, 최종적으로 출력하는 activation맵의 차원은 segmentation하고자 하는 class의 개수와 같다(softmax-crossentropy loss 사용시) 가령, 흑백으로 mask가 주어지면, 채널이 2인 것이다. UNet의 주목할만한 특징은 'skip connection'이다. 이는 인코더의 activation map을 대응하는 디코더의 activation map과 concat하는 것으로, 원본이미지의 비교적 정확한 위치정보를 디코더에서 건네받는 셈으로, 정확한 위치 복원에 도움이 되고 결과론적으로 segmentation 결과도 좋아지게 된다. 

```python
#UNet에서 반복되는 구조를 함수로 표현

def conv_block(in_dim, out_dim, act_fn):
    model = nn.Sequential(nn.Conv2d(in_dim, out_dim, kernel_size=3, stride=1, padding=1),
                          nn.BatchNorm2d(out_dim),
                          act_fn)
    return model

def conv_trans_block(in_dim, out_dim, act_fn):
    model = nn.Sequential(nn.ConvTranspose2d(in_dim, out_dim, kernel_size=3, stride=2, padding=1, output_padding=1),
                          nn.BatchNorm2d(out_dim),
                          act_fn)
    return model

def maxpool():#사이즈 절반으로 줄임
    pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
    return pool

def conv_block_2(in_dim, out_dim, act_fn): #사이즈는 변하지 않음, 채널크기만 변함.
    model = nn.Sequential(conv_block(in_dim, out_dim, act_fn),
                          nn.Conv2d(out_dim, out_dim, kernel_size=3, stride=1, padding=1),
                          nn.BatchNorm2d(out_dim))
    return model
```

```python
class Unet(nn.Module):
    def __init__(self, in_dim, out_dim, num_filter):
        super(Unet, self).__init__()
        self.in_dim = in_dim
        self.out_dim = out_dim
        self.num_filter = num_filter
        act_fn = nn.LeakyReLU(0.2, inplace=True)
        print("\n-----------------Initiating U-Net---------------\n")
        
        self.down_1 =  conv_block_2(self.in_dim, self.num_filter, act_fn)
        self.pool_1 = maxpool() #1/2
        self.down_2 = conv_block_2(self.num_filter, self.num_filter*2, act_fn)
        self.pool_2 = maxpool() #1/2
        self.down_3 = conv_block_2(self.num_filter*2, self.num_filter*4, act_fn)
        self.pool_3 = maxpool() #1/2
        self.down_4 = conv_block_2(self.num_filter*4, self.num_filter*8, act_fn)
        self.pool_4 = maxpool() #1/2
        
        self.bridge = conv_block_2(self.num_filter*8, self.num_filter*16, act_fn)
        
        self.trans_1 = conv_trans_block(self.num_filter*16, self.num_filter*8, act_fn) # x2
        self.up_1 = conv_block_2(self.num_filter*16, self.num_filter*8, act_fn) #channel이 down_4를 거친 map과 concat되므로
        self.trans_2 = conv_trans_block(self.num_filter*8, self.num_filter*4, act_fn) #x2
        self.up_2 = conv_block_2(self.num_filter*8, self.num_filter*4, act_fn) #down3과 concat
        self.trans_3 = conv_trans_block(self.num_filter*4, self.num_filter*2, act_fn)#x2
        self.up_3 = conv_block_2(self.num_filter*4, self.num_filter*2, act_fn) #down2와 concat
        self.trans_4 = conv_trans_block(self.num_filter*2, self.num_filter, act_fn) #x2
        self.up_4 = conv_block_2(self.num_filter*2, self.num_filter, act_fn) #down1과 ooncat
        self.out = nn.Sequential(nn.Conv2d(self.num_filter, self.out_dim, 3, 1, 1),
                                 nn.Tanh())
        
    def forward(self, x):
        down1 = self.down_1(x)
        pool1 = self.pool_1(down1)
        down2 = self.down_2(pool1)
        pool2 = self.pool_2(down2)
        down3 = self.down_3(pool2)
        pool3 = self.pool_3(down3)
        down4 = self.down_4(pool3)
        pool4 = self.pool_4(down4)
        
        bridge = self.bridge(pool4)
        
        trans1 = self.trans_1(bridge)
        concat_1 = torch.cat([trans1, down4], dim=1)
        up1 = self.up_1(concat_1)
        trans2 = self.trans_2(up1)
        concat_2 = torch.cat([trans2, down3], dim=1)
        up2 = self.up_2(concat_2)
        trans_3 = self.trans_3(up2)
        concat_3 = torch.cat([trans_3, down2], dim=1)
        up3 = self.up_3(concat_3)
        trans_4 = self.trans_4(up3)
        concat_4 = torch.cat([trans_4, down1], dim=1)
        up4 = self.up_4(concat_4)
        out = self.out(up4)
        return out
```

```python
model = Unet(3, 2, 16)
```

```python
input_img = torch.Tensor(24, 3, 256, 256) #모델 테스트용 이미지 생성
output = model(input_img)
```
